```python
import numpy as np
import pandas as pd
```

```python
pgo = pd.read_csv('dmspgoout.log')
pgo['frequency'] = pgo['Eupper']-pgo['Elower']
pgo['Intensity'] = pgo['Intensity']/np.max(pgo['Intensity'])
pgo.sort_values('frequency', ascending=True, inplace=True)
display(pgo)
print(len(pgo))
```

```python
me = pd.read_csv('spectrum.csv')
me['Intensity'] = me['intensity']/np.max(me['intensity'])
display(me)
print(len(me))
```

```python

tol = 1e-5


f1 = pgo[['frequency', 'Intensity']].values
f2 = me[['frequency', 'Intensity']].values

pgocomp = np.isclose(f1[:, None, :], f2[None, :, :], atol=tol)

pgom = np.all(pgocomp, axis=2)


mask1 = ~np.any(pgom, axis=1)
upgo = pgo[mask1]

mecomp = np.isclose(f2[:, None, :], f1[None, :, :], atol=tol)
mem = np.all(mecomp, axis=2)
mask2 = ~np.any(mem, axis=1)
ume = me[mask2]

print("Unique to pgo:\n", upgo)
print("Unique to me:\n", ume)
```
